```python
import pandas as pd

# Load the dataset
Tdata = pd.read_csv('D:/DataAnalytics/MyProjects/Titanic-Dataset.csv')

# Step 2: Check for missing values
print("Missing values before cleaning:")
print(Tdata.isnull().sum())

# Option 1: Fill missing values
# Fill missing 'Age' with the median age
Tdata['Age'] = Tdata['Age'].fillna(Tdata['Age'].median())

# Fill missing 'Embarked' with the most common value
Tdata['Embarked'] = Tdata['Embarked'].fillna(Tdata['Embarked'].mode()[0])

# Option 2: Drop columns with too many missing values, like 'Cabin'
if 'Cabin' in Tdata.columns:
    Tdata.drop(columns=['Cabin'], inplace=True)

# Recheck for missing values
print("\nMissing values after cleaning:")
print(Tdata.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a histogram for the 'Age' column
plt.figure(figsize=(10, 6))
sns.histplot(Tdata['Age'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create a bar chart for survival count
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', data=Tdata)
plt.title('Count of Survived vs Not Survived')
plt.xlabel('Survived (1 = Yes, 0 = No)')
plt.ylabel('Count')
plt.show()
```

```python
# Bar plot for survival rate by gender
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', hue='Sex', data=Tdata)
plt.title('Survival Rate by Gender')
plt.xlabel('Survived (1 = Yes, 0 = No)')
plt.ylabel('Count')
plt.show()
```

```python
# Bar plot for survival rate by passenger class
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', hue='Pclass', data=Tdata)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Survived (1 = Yes, 0 = No)')
plt.ylabel('Count')
plt.show()
```

```python
# Group survival by both 'Sex' and 'Pclass'
plt.figure(figsize=(10, 6))
sns.catplot(x="Pclass", hue="Sex", col="Survived", data=Tdata, kind="count", height=4, aspect=.7)
plt.show()
```
